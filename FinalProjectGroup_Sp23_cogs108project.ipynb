# Research Question

How do people’s health behaviors affect the healthy degree of their weight in the U.S. in 2017? 

- Health behavior is defined as people's intentional actions to improve their health. In this context, we consider people's behaviors but not their instinctly healthy conditions (for example, having health insurance and doing exercise regularly are health behaviors, but genetical illness are not). Conversely, actions that are detrimental to health imply a lack of health behavior (for example, smoking or drinking).The way we quantize will be discussed below. 

- The healthy degree of one's weight is defined based on the standard BMI scale, and it indicates how healthy one’s weight is. For the BMI standard, the healthy range is [18.5, 25). With a BMI out of this range, one is considered to be underweight or overweight, which is less healthy. If one’s BMI is larger than 30, we will consider one obese, which is morbid and should be treated as a disease. We will talk more about how we use BMI to measure this variable in the following parts.

### Assessment of Missingness 

This dataset uses certain values to represent missingness (see specific values description for each variable in the variable table), so we cannot simply use dropna() in pandas. Instead, we need to turn the invalid values into NaN first.

```python
## set NaN values according to value codes

# BMI
data.loc[(data['BMI'] == 0) | (data['BMI'] >= 99),'BMI'] = np.nan

# lacks coverage
data.loc[data['HINOTCOVE'] == 9, 'HINOTCOVE'] = np.nan

# no insurance
data.loc[data['HINONE'] >= 7, 'HINONE'] = np.nan

# insurance cost
data.loc[(data['HIP1COST'] >= 99990), 'HIP1COST'] = np.nan

# number of drinking days
data.loc[data['ALC5UPYR'] > 365, 'ALC5UPYR'] = np.nan

# cigarette per day
data.loc[data['CIGDAYMO'] >= 90, "CIGDAYMO"] = np.nan

# quit smoking number and units
data.loc[data['QUITNO'] >= 900, 'QUITNO'] = np.nan
data.loc[data['QUITTP'] >= 6, 'QUITTP'] = np.nan

# smoke frequency
data.loc[data['SMOKFREQNOW'] >= 5, 'SMOKFREQNOW'] = np.nan

# physical activities variables
def clean_physical_activities(val):
    if val >= 97:
        return np.nan
    elif val >= 94:
        return 0
    else:
        return val
    
data['MOD10FWK'] = data['MOD10FWK'].apply(clean_physical_activities)
data['VIG10FWK'] = data['VIG10FWK'].apply(clean_physical_activities)
data['STRONGFWK'] = data['STRONGFWK'].apply(clean_physical_activities)

# loss weight related variables
def clean_weight_loss(val):
    if val == 0 or val >= 5:
        return np.nan
    else:
        return val
    
data['DIETDYR'] = data['DIETDYR'].apply(clean_weight_loss)
data['WTPROGDYR'] = data['WTPROGDYR'].apply(clean_weight_loss)
data['DIETNOW'] = data['DIETNOW'].apply(clean_weight_loss)
data['WTPROGNOW'] = data['WTPROGNOW'].apply(clean_weight_loss)

# search health info online
data['PCLOOKHELYR'] = data['PCLOOKHELYR'].apply(clean_weight_loss)

# sleep time per day
data.loc[(data['HRSLEEP'] > 24) | (data['HRSLEEP'] == 0), 'HRSLEEP'] = np.nan

data.head()
```

### Construct Variables

- **health behavior index**: First, we categorizing the health-behavior-related variables to beneficial/unbeneficial, depending on their effects on one's health are positive or negative. Then we sum them up to get the health behavior index. Higher its value is, more health behaviors the individual is considered to have. 
- **healthy degree of weight**: In a categorical way, we build it by reassigning values to the categorical BMI: 1=healthy, 0=not healthy, -1=morbid. In the numerical way, we use the BMI variable and set the center of healthy range (21.75) to have a "healthy score" of 50. So this score will be 50 - (absolute distance to 21.75). 

See the comments in each cell for what's going.

```python
df_hbi = data_cl.drop(columns=['REGION', 'NHISPID', 'SAMPWEIGHT','AGE','SEX','BMI','BMICAT'])
```

```python
df_hbi.describe()
```

```python
# First, normalize some numerical variables which are too big.
# Divide them by the mean.
df_hbi['HIP1COST'] = df_hbi['HIP1COST'] / df_hbi['HIP1COST'].mean()
df_hbi['ALC5UPYR'] = df_hbi['ALC5UPYR'] / df_hbi['ALC5UPYR'].mean()
df_hbi['MOD10FWK'] = df_hbi['MOD10FWK'] / df_hbi['MOD10FWK'].mean()
df_hbi['VIG10FWK'] = df_hbi['VIG10FWK'] / df_hbi['VIG10FWK'].mean()
df_hbi['STRONGFWK'] = df_hbi['STRONGFWK'] / df_hbi['STRONGFWK'].mean()
```

```python
# construct variable: days of quit smoking
def quit_smk(time, unit):
#     2=week, 3=month, 4=year
    unit = unit.replace({2:7, 3:30, 4:365})
    out = time * unit
    return out

data['quit_smoke_days'] = quit_smk(data['QUITNO'], data['QUITTP'])
df_hbi['quit_smk_days'] = data['quit_smoke_days'] / data['quit_smoke_days'].mean()
df_hbi = df_hbi.drop(columns=['QUITNO', 'QUITTP'])
```

For sleep hours, we set 7 hours as the most healthy (value = 1)<sup>[Dimitriu]</sup>. 

<sup>[Dimitriu]: Dimitriu, A. (2020, September 18). 7 hours of sleep: The New Gold Standard?. Psychology Today. https://www.psychologytoday.com/us/blog/psychiatry-and-sleep/202009/7-hours-sleep-the-new-gold-standard </sup>

```python
# construct variable: healthy degree of sleep hours
sleep = df_hbi['HRSLEEP'] / df_hbi['HRSLEEP'].mean()
sleep = 1 - abs(sleep - 1)
df_hbi = df_hbi.drop(columns=['HRSLEEP'])
df_hbi['sleep_health'] = sleep
```

```python
# calculate HBI by adding things up
def HBI(df):
    """
    input: a dataframe with all variables needed
    output: a series of HBI (health behavior index)
    """
    out = df.sum(axis=1)
    return out

data_cl['HBI'] = HBI(df_hbi)
```

```python
# define the healthy degree of weight using method mentioned above
# numerical
data_cl['weight_healthy_deg'] = 50 - abs(21.75 - data['BMI'])

# categorical
# 1=underweight, 2=normal, 3=overweight, 4=obese for BMICAT
dct = {1:0, 3:0, 2:1, 4:-1}
data_cl['weight_healthy_deg_cate'] = data_cl['BMICAT'].replace(dct)
```

```python
# renaming the columns for readability
data_cl = data_cl.rename(columns = {"REGION":"region","NHISPID":"PID","SAMPWEIGHT":"sample_weight","AGE":'age','SEX':'sex',
                                    "BMICAT":"BMI_cate","HINOTCOVE":"lack_HI_cover","HINONE":"no_HI",
                                    "HIP1COST":"fee_HI","ALC5UPYR":"drinks_more_than_5",
                                    "SMOKFREQNOW":"smoke_freq", "SMKLSFREQNOW":"tobacco_freq","QUITNO":"quit_smoke_time",
                                    "QUITTP":"unit_quit_smoke_time","MOD10FWK":"moderate_physc_act",
                                    "VIG10FWK":"vigorous_physc_act","STRONGFWK":"strengthen_act",
                                    "DIETDYR":"told_diet","WTPROGDYR":"told_weight_loss",
                                    "DIETNOW":"now_diet","WTPROGNOW":"now_weight_loss","HRSLEEP":"hours_sleep",
                                    "PCLOOKHELYR":"search_health_info_online"})
data_cl
```

```python
def find_outliers_IQR(df):
    q1 = df.quantile(0.25)
    q3 = df.quantile(0.75)
    IQR = q3-q1
    outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]
    return outliers


def drop_outliers_IQR(df):
    q1 = df.quantile(0.25)
    q3 = df.quantile(0.75)
    IQR = q3-q1
    outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]
    
    not_outliers = df[~((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]
    outliers_dropped = not_outliers.dropna().reset_index()
    return outliers_dropped

outliers_HBI = find_outliers_IQR(data_cl['HBI'])
outliers_BMI = find_outliers_IQR(data_cl['BMI'])
outliers_degree = find_outliers_IQR(data_cl['weight_healthy_deg'])
print(outliers_HBI)
print('\n')
print(outliers_BMI)
print('\n')
print(outliers_degree)
```

### Simple OLS

First let's take a look on a simple linear regression model using OLS estimator. 

Consider the assumptions of this model, our data seems not really normally distributed or homoscedastic. To solve this, let's use log transformation of treatment variable `HBI`. The outcome variable `weight_healthy_deg` is not likely to be normally distributed due to its design, and taking log only makes it less normal. But the underlying `BMI` is normal.

```python
df['HBI_log'] = df['HBI'].apply(np.log1p)

# For weight_healthy_degree < 0, assign them with -1 simply
df['weightheal_log'] = df['weight_healthy_deg'].apply(np.log1p)
df.loc[df['weightheal_log'].isnull(), 'weightheal_log'] = -1
```

```python
plt.figure(figsize=(25,10))

plt.subplot(121)
sns.distplot(df['HBI_log'])

plt.subplot(122)
sns.distplot(df['HBI']);
```

```python
plt.figure(figsize=(25,10))

plt.subplot(121)
sns.scatterplot(x='HBI', y='weight_healthy_deg', data=df)
plt.xlabel('Original Data')

plt.subplot(122)
sns.scatterplot(x='HBI_log', y='weight_healthy_deg', data=df)
plt.xlabel('Log-transformed Data')
```

We can see that HBI is distributed in a more normal way after log transformation. Also, the heteroscedasticity problem is improved a lot. 

```python
# perform simple linear regression
out1, preds1 = patsy.dmatrices("weight_healthy_deg ~ HBI_log", df)
mod1 = sm.OLS(out1, preds1)
res1 = mod1.fit()
print(res1.summary())
```

### Simple WLS

Apply sample weight in the model to deal with sampling bias and heteroscedasticity.

```python
# weighted simple linear regression
out1_w, preds1_w = patsy.dmatrices("weight_healthy_deg ~ HBI_log", df)
mod1_w = sm.WLS(out1_w, preds1_w, weights=df['sample_weight'])
res1_w = mod1_w.fit()
print(res1_w.summary())
```

### Multiple OLS

Add age and sex as control variables.

```python
# multiple linear regression
# data processing: one-hot encode for sex
df['male'] = (df['sex'] == 1).astype(int)

# perform
out2, preds2 = patsy.dmatrices("weight_healthy_deg ~ HBI_log + age + male", df)
mod2 = sm.OLS(out2, preds2)
res2 = mod2.fit()
print(res2.summary())
```

### Multiple WLS

Apply sample weights on multiple linear regression model.

```python
# using sample weight on multiple lin-reg
out2_w, preds2_w = patsy.dmatrices("weight_healthy_deg ~ HBI_log + age + male", df)
mod2_w = sm.WLS(out2_w, preds2_w, weights=df['sample_weight'])
res2_w = mod2_w.fit()
print(res2_w.summary())
```

### Consider more - Backward Stepwise Regression

We suspect that some health behavior variables affect the whole performance of health behavior index. To figure out if there are any reliable regressors, let's try stepwise regression.

```python
# modify behavior columns
df = data_cl.copy().drop(columns=['region', 'PID', 'sample_weight', 'BMI', 'BMI_cate',
                          'HBI', 'weight_healthy_deg_cate'])

df['quit_smoke_days'] = quit_smk(df['quit_smoke_time'], df['unit_quit_smoke_time'])
df['quit_smoke_days'] = df['quit_smoke_days'] / df['quit_smoke_days'].mean()
df = df.drop(['quit_smoke_time', 'unit_quit_smoke_time'], axis=1)

sleep = df['hours_sleep'] / df['hours_sleep'].mean()
sleep = 1 - abs(sleep - 1)
df = df.drop(columns=['hours_sleep'])
df['sleep_health'] = sleep

y = df['weight_healthy_deg']
X = df.drop('weight_healthy_deg', axis=1)
X = sm.add_constant(X)
```

```python
# try to not aggregate HBI, stepwise
model = sm.OLS(y, X)

# Fit the model
results = model.fit()

# Perform backward stepwise regression
variables = list(X.columns)  # variable names
p_value_threshold = 0.01

while len(variables) > 1:  # removing variables until only constant left
    # fit with the current set
    X_subset = X[variables]
    model = sm.OLS(y, X_subset)
    results = model.fit()
    
    # get the p-values
    p_values = results.pvalues.iloc[1:]
    
    # find the variable with the highest p-value, check threshold
    variable_to_remove = p_values.idxmax()
    max_p_value = p_values.max()
    if max_p_value > p_value_threshold:
        # exceed => remove
        variables.remove(variable_to_remove)
    else:
        # all remaining variables are significant!
        break  

# Fit the final model
X_final = X[variables]
model = sm.OLS(y, X_final)
final_results = model.fit()
```
